### Notch filter

```python
SAMPLING_FREQUENCY = 256
POWER_LINE_FREQUENCY = 50

#########################################3

filter_bank = FilterBank(SAMPLING_FREQUENCY)
frequency_range, h = filter_bank.notch(POWER_LINE_FREQUENCY, get_freqz=True)

plot_frequency_and_phase_response(frequency_range, h)
```

### High pass butterworth

```python
SAMPLING_FREQUENCY = 256
CUT_FREQUENCY = 0.5

#########################################

filter_bank = FilterBank(SAMPLING_FREQUENCY)
frequency_range, h = filter_bank.highpass_butter(CUT_FREQUENCY, get_freqz=True)

plot_frequency_and_phase_response(frequency_range[:20], h[:20])
```
