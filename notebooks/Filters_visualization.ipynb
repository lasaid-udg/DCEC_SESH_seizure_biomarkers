```python
import sys
import numpy
sys.path.append("../")
from lib.filters import FilterBank
from lib.visuals import plot_frequency_and_phase_response
```

### Notch filter

```python
SAMPLING_FREQUENCY = 256
POWER_LINE_FREQUENCY = 50

#########################################3

filter_bank = FilterBank(SAMPLING_FREQUENCY)
frequency_range, h = filter_bank.notch(POWER_LINE_FREQUENCY, get_freqz=True)

plot_frequency_and_phase_response(frequency_range, h)

#########################################
for idx in range(195, 205):
    ratio = numpy.abs(h[idx])
    gain = 20*numpy.log10(ratio)
    print(f"freq={frequency_range[idx]}Hz, \t ratio={ratio}, \t gain={gain}dB")
```

### High pass butterworth

```python
SAMPLING_FREQUENCY = 256
CUT_FREQUENCY = 0.5

#########################################

filter_bank = FilterBank(SAMPLING_FREQUENCY)
frequency_range, h = filter_bank.highpass_butter(CUT_FREQUENCY, get_freqz=True)

plot_frequency_and_phase_response(frequency_range[:20], h[:20])

#########################################
for idx in range(10):
    ratio = numpy.abs(h[idx])
    gain = 20*numpy.log10(ratio)
    angle = numpy.angle(h[idx])*180/numpy.pi
    print(f"freq={frequency_range[idx]}Hz, \t ratio={ratio}, \t gain={gain}dB, \t phase-shift={angle}Â°")
```

### Low pass butterworth

```python
SAMPLING_FREQUENCY = 256
CUT_FREQUENCY = 80

#########################################

filter_bank = FilterBank(SAMPLING_FREQUENCY)
frequency_range, h = filter_bank.lowpass_butter(CUT_FREQUENCY, get_freqz=True)

plot_frequency_and_phase_response(frequency_range, h)

#########################################
for idx in range(312, 322):
    ratio = numpy.abs(h[idx])
    gain = 20*numpy.log10(ratio)
    print(f"freq={frequency_range[idx]}Hz, \t ratio={ratio}, \t gain={gain}dB")
```

###  Band pass butterworth

```python
SAMPLING_FREQUENCY = 256
CUT_FREQUENCY = (4, 8)

#########################################

filter_bank = FilterBank(SAMPLING_FREQUENCY)
frequency_range, h = filter_bank.bandpass_butter(CUT_FREQUENCY, 4, get_freqz=True)

plot_frequency_and_phase_response(frequency_range, h)

#########################################
for idx in range(10, 30):
    ratio = numpy.abs(h[idx])
    gain = 20*numpy.log10(ratio)
    print(f"freq={frequency_range[idx]}Hz, \t ratio={ratio}, \t gain={gain}dB")
```
